# Train test split

```python
Y = df["status"]
X = df.drop(["status"],axis = 1)
Y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,Y,test_size=0.30)
```

## ML MODELS

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
```

```python
lor.fit(X_train, y_train) #input and output of training
#model.get_params()
print("Accuracy: ",lor.score(X_test,y_test)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dclf = DecisionTreeClassifier()
```

```python
dclf = dclf.fit(X_train,y_train)
dclf.fit(X_train,y_train)
```

```python
dclf.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfclf = RandomForestClassifier(n_estimators=100)
rfclf = rfclf.fit(X_train,y_train)
print(rfclf.score(X_test,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train) #input and output of training
#model.get_params()
print("Accuracy: ",knn.score(X_test,y_test)*100)
```
